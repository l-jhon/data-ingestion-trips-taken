## Reading CSV to analyze data

```python
df_trips = pd.read_csv('../data/trips.csv')
df_trips.head()
```

## Get latitude and longitude from origin_coord and destination_coord

```python
df_trips["origin_coord_lat"] = df_trips["origin_coord"].str.replace("POINT|[()]", "", regex=True).str.split(" ").apply(lambda x: x[1])
df_trips["origin_coord_lng"] = df_trips["origin_coord"].str.replace("POINT|[()]", "", regex=True).str.split(" ").apply(lambda x: x[2])
df_trips["destination_coord_lat"] = df_trips["destination_coord"].str.replace("POINT|[()]", "", regex=True).str.split(" ").apply(lambda x: x[1])
df_trips["destination_coord_lng"] = df_trips["destination_coord"].str.replace("POINT|[()]", "", regex=True).str.split(" ").apply(lambda x: x[2])
```
